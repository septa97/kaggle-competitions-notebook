## Import libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.model_selection import train_test_split

from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers.convolutional import Conv2D, MaxPooling2D
from keras.utils import np_utils
from keras import backend as K

# Use TensorFlow dimension ordering (rows, columns, depth). If using Theano, use (depth, rows, columns) instead
K.set_image_dim_ordering('tf')
```

## Import CSV files

```python
data_df = pd.read_csv('train.csv')
input_df = pd.read_csv('test.csv') # The input data that needs to be submitted to Kaggle
```

```python
data_df.describe()
```

## Data preprocessing

```python
X_train = data_df.loc[:, data_df.columns != 'label'].values
y_train = data_df['label'].values

X_train = X_train.reshape(X_train.shape[0], 28, 28, 1).astype('float32') / 255
y_train = np_utils.to_categorical(y_train, 10) # Set the labels as categorical
```

## Train the classifier

```python
model = Sequential()

model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(Conv2D(32, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, batch_size=32, epochs=30, verbose=1)
```

```python
train_score = model.evaluate(X_train, y_train, verbose=0)
print(train_score)
```

```python
input_matrix = input_df.values.reshape(input_df.values.shape[0], 28, 28, 1)
input_prediction_labels = np.argmax(model.predict(input_matrix), axis=1)
```

```python
result_df = pd.DataFrame({
    'ImageId': pd.Series(np.arange(1, input_df.shape[0]+1)),
    'Label': pd.Series(input_prediction_labels)
})

result_df.to_csv('result.csv', index=False)
```
